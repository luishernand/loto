```python
# Make a funtion to scrape webside
def scrape_web(page_date):
    #page_num = str(page_date)
    url = 'https://www.conectate.com.do/loterias/leidsa/loto-mas?date='+ page_date
    webpage = requests.get(url)
    soup = BeautifulSoup(webpage.text, 'html.parser')
    resultados = soup.find_all('span', attrs={'class': 'score'})
    fecha = soup.find_all('span', attrs={'class': 'session-date session-badge'})
    
    num = []
    dates = []
    for i in resultados:
        numero = i.parent.text.strip().split('\n') # Get the quoute text  numbers
        num.append(numero)
    
    for i in fecha:
        date = i.parent.text.strip().split('\n')[0] # Get the quoute text fecha
        dates.append(date)
        
       
    return dates, num
```

```python
fechas, num = scrape_web('18-08-2021')
```
